```python
import json
import os
import glob

import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

# 데이터 불러오기

```python
root_path = '/content/drive/MyDrive/wjdckdtn15@hashscraper.com/HASHNLP/data/category_classifier/국립국어원 신문 말뭉치 2021(버전 1.0)'
target_path = os.path.join(root_path+'/*.json')
```

```python
target_path
```

```python
path_list = []
for path in glob.glob(target_path):
  path_list.append(path)

path_list
```

```python
title, content, label = [], [], []
for path in path_list:
  with open(path) as f:
    data = json.load(f)
    data = data['document']
    for datum in data:
      title.append(datum['paragraph'][0]['form'])
      content.append(" ".join(sentence['form'] for sentence in datum['paragraph'][1:]))
      label.append(datum['metadata']['topic'])

my_data = pd.DataFrame({'title': title,
                        'content': content,
                        'label': label})

my_data
```

```python
my_data['label'].value_counts()
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(my_data['label'])
num_labels = len(label_encoder.classes_)

my_data['label'] = np.asarray(label_encoder.transform(my_data['label']), dtype=np.int32)
my_data['label'] = my_data['label'].apply(lambda x: int(x))
my_data
```

```python
my_data['label'].value_counts()
```

```python
my_data.to_csv('/content/drive/MyDrive/wjdckdtn15@hashscraper.com/HASHNLP/data/category_classifier/news_category(nikl).csv', index=False)
```

# 언더 샘플링

```python
new_data = pd.DataFrame()
for label in range(0, num_labels):
  one_category = my_data[my_data['label'] == label].sample(6979, random_state=42)
  new_data = pd.concat([new_data, one_category])

new_data = new_data.sample(frac=1, random_state=42).reset_index(drop=True)
new_data
```

# 타이틀만

```python
title_data = new_data.drop('content', axis=1)
title_data
```

```python
title_data.columns = ['sentence', 'label']
```

```python
train_title_data, test_title_data = train_test_split(title_data, test_size=0.2, random_state=42)
```

```python
val_title_data, test_title_data = train_test_split(test_title_data, test_size=0.5, random_state=42)
```

```python
train_title_data.to_csv('/content/drive/MyDrive/wjdckdtn15@hashscraper.com/HASHNLP/data/category_classifier/train_title_data.csv', index=False)
val_title_data.to_csv('/content/drive/MyDrive/wjdckdtn15@hashscraper.com/HASHNLP/data/category_classifier/val_title_data.csv', index=False)
test_title_data.to_csv('/content/drive/MyDrive/wjdckdtn15@hashscraper.com/HASHNLP/data/category_classifier/test_title_data.csv', index=False)
```
