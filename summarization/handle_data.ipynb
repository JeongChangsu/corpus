```python
root_path = 'VL1'
target_path = os.path.join(root_path+'/*')

path_list = []
for path in glob.glob(target_path):
  path_list.append(path)

path_list
```

```python
path_list2 = []
for path in path_list:
    real_target_path = os.path.join(path+'/2~3sent/*.json')
    for real_path in glob.glob(real_target_path):
        path_list2.append(real_path)
        
path_list2
```

```python
text, label = [], []
for path in path_list2:
    with open(path) as f:
        data = json.load(f)
        text.append(data['Meta(Refine)']['passage'])
        label.append(data['Annotation']['summary1'])

my_data = pd.DataFrame({'text': text,
                       'label': label})

my_data
```

```python
test = my_data
```

```python
my_data = pd.concat([train, test])
my_data
```

```python
my_data.duplicated('text').value_counts()
```

```python
my_data.drop_duplicates('text', inplace=True)
my_data
```

```python
my_data = my_data.sample(frac=1, random_state=42)
my_data
```

```python
my_data.reset_index(drop=True, inplace=True)
my_data
```

```python
my_data.isna().sum()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(my_data, test_size=0.1, random_state=42)
```
